### Aggregating
#### Parse out date from time variable

```python
sepsis['new_date'] = sepsis['time'].str[:9]
sepsis.head(15)
```

#### Make unique ID to group by

```python
sepsis['unique_id'] = sepsis['patient_id'].map(str) + sepsis['encounter_id'].map(str) + sepsis['new_date']
sepsis.head(5)
```

#### Aggregating

```python
aggregations = {
    'avpu':max,
    'avpu_old':max,
    'temp':max,
    'staph':max,
    'age_at_enc':'first',
    'bmi':'first',
    'race':'first',
    'sex':'first',
    'ethnicity':'first',
    'treatment_limit':max,
    'dnr':max,
    'dnr_dni':max,
    'dncpr_dni':max,
    'dnr_treatment_limit':max,
    'pain_scale': max,
    'albuterol': max,
    'immunosupp_medname':max,
    'immunosupp_class':max,
    'biologicals':max,
    'icd9_477_x':max,
    'icd9_691_x':max,
    'icd9_493_x':max,
    'icd9_995_3':max,
    'icd9_558_3':max,
    'antibiotic':max,
    'sepsis_glucocorticoid':max,
    'sepsis_antibiotic':max,
    'infxnqsofa':max,
    }
```

```python
sepsis_group = sepsis.groupby(['unique_id']).agg(aggregations)
list(sepsis_group)
```
